```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
match_data = pd.read_csv('IPL Matches.csv')
match_data.head()
```

```python
#match with the bigest win by run margin
biggest_win_by_runs = match_data['result_margin'].max()
match_detail=match_data.loc[match_data['result_margin']==biggest_win_by_runs]
match_detail[['city','date','player_of_match','venue','team1','team2','winner','result_margin']]
```

```python
# match with the highest win margin by wkts 
biggest_by_wkts = (match_data['result_margin'][match_data['result']=='wickets']).max()
match_detail=match_data.loc[(match_data['result_margin']==biggest_by_wkts)&(match_data['result']=='wickets')]
match_detail[['city','date','player_of_match','venue','team1','team2','winner','result_margin']]
```

```python
#matches which got tie
tie=match_data.loc[match_data['result']=='tie']
venue_with_most_tie=(tie['venue'][tie['result']=='tie']).value_counts()
tie[['city','date','player_of_match','venue','team1','team2','winner']]
```

```python
#matches win by dl(Duckworth-lewis Method), team that win by most dl and there venues
win_by_dl=(match_data.loc[match_data['method']=='D/L'])
team_with_most_dl=(win_by_dl['winner'][win_by_dl['method']=='D/L']).value_counts()
venue_with_most_dl=(win_by_dl['venue'][win_by_dl['method']=='D/L']).value_counts()
win_by_dl[['city','venue','team1','team2','winner']]
```

```python
print("Team with most matches win by d/l system :- ")
team_with_most_dl
```

```python
#Man of the match
man_of_the_match=match_data['player_of_match'].value_counts().sort_values(ascending=False)
df=pd.DataFrame(man_of_the_match.values,man_of_the_match.index, columns=['No. of Man of the match'])
df.index.names=['Player']
df
```

```python
#popular venues and number of matches played there
venue_with_most_matches=(match_data['venue']).value_counts()
df=pd.DataFrame(venue_with_most_matches.values,venue_with_most_matches.index, columns=['No. of the match played'])
df.index.names=['Venue']
df[:7]
# venue_with_most_matches
```

```python
Eden_Gardens=match_data['venue'][match_data['venue']=='Eden Gardens']
win_by_run = match_data['result_margin'][match_data['result']=='runs'][match_data['venue']=='Eden Gardens']
slices=[len(win_by_run),(len(Eden_Gardens)-len(win_by_run))]
label=['Batting first','Batting second']
plt.pie(slices,labels=label,startangle=90,shadow=1,explode=(0,0.4),autopct='%1.2f%%',textprops={'size': 'x-large'})
plt.show()
```
